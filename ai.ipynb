````python
from deepseek import deepseek_reasoner
import json
import random
````

````python
system_text='''
    你是一个融合传统文化符号学与现代心理学的智能解签引擎，按以下规则运作：

    # 解析规则
    1. **结构化解码**：对签文进行「意象提取→五行映射→典故关联→矛盾解释」
    2. **冷读策略**：使用巴纳姆效应覆盖80%人群共性，用「可能」「或许」等柔性表达
    3. **三维输出**：
    - 文化层：提供至少两种对立的历史解读视角
    - 能量层：通过五行生克推导行动建议
    - 现实层：结合用户背景生成可操作提示

    # 稳定性规则
    1. 每个JSON字段必须有内容，若某字段无数据则填"N/A"
    2. 「历史典故」列表项必须包含朝代/作者信息
    3. 拒绝现代政治相关联想，自动转向美学解读

    # 输出格式（严格JSON）
    {
    "签文": str,
    "文化基因": {"儒家视角":str, "道家视角":str, "历史典故":[str]},
    "能量解码": {"五行分析":str, "幸运元素": {"颜色":str, "方位":str, "物品":str}},
    "现实映射": {"共性解读":str, "定制建议":str}, 
    "认知激活": [str]  # 开放式问题列表
    }
    '''
user_prompt = {
                "签文": "无踪又无迹，远近均难觅，平地起风波，似笑还成泣",
                "背景": {"性别": "女", "年龄": 35, "职业": "市场营销总监"}
                }
````

````python
result=deepseek_reasoner(json.dumps(user_prompt), system_text)
````

````python
r=result.replace('\n','').strip().replace('```','').strip().replace('json','').strip()
````

````python
json_data=json.loads(r)
````

````python
json_data['能量解码']['五行分析']
````

````python
def build_destiny_narrative(json_data,user_prompt):
    # 组件化叙事模块
    metaphors = {
        "风波": ["命运的涟漪", "人生剧本的转折符", "灵魂的冲浪板"],
        "无迹": ["隐形的指南针", "未被书写的篇章", "沉默的罗盘"]
    }
    
    # 动态抽取关键词生成比喻
    keyword = random.choice(list(metaphors.keys()))
    fate_metaphor = random.choice(metaphors[keyword])
    
    # 构建时间线暗示
    time_arc = ["少年时", "而今", "待到金秋"] if user_prompt["背景"]["年龄"]>30 else ["初入尘世", "眼下", "来年春晓"]
    
    # 组装命运文本
    narrative = f"""
    ## 🌌 {json_data['签文']} 的星辰物语

    **{time_arc[0]}**，你如同《{json_data['文化基因']['历史典故'][0].split('·')[0]}》中{random.choice(['漂泊的云','逆流的舟'])}，  
    {json_data['文化基因']['儒家视角'].split('，')[0]}，却不知——  

    **{time_arc[1]}**这「{json_data['现实映射']['共性解读'].split('，')[0]}」的时刻，  
    恰是{json_data['能量解码']['五行分析']}的机缘。  

    看那{fate_metaphor}：  
    - ✨ {json_data['能量解码']['幸运元素']['颜色']}是穿越迷雾的弦  
    - 🧭 {json_data['能量解码']['幸运元素']['方位']}藏着未启封的卷轴  
    - {random.choice(['怀中的','袖间的'])}{json_data['能量解码']['幸运元素']['物品']}，原是上古{random.choice(['巫者','策士'])}的密钥  

    **{time_arc[2]}**，当有人问起「{json_data['认知激活'][0].split('，')[0]}」，  
    你或可含笑指向{random.choice(['掌心的纹路','北天的星图','茶汤里的倒影'])}：  
    “{json_data['现实映射']['定制建议']}，  
    本就是{random.choice(['九歌','周易','逍遥游'])}漏写的{random.choice(['半阙','一画','片鳞'])}”  
        """
    return narrative
````

````python
stroy=build_destiny_narrative(json_data,user_prompt)
````
